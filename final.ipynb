```python
# BFS, plan a subway route
def plan_path(starting_point, destination):
    if starting_point == destination: 
        return [starting_point] 
    explored = set()
    queue = [ [starting_point] ]
    while queue: 
        path = queue.pop(0) 
        s = path[-1] 
        for state, action in beijing_subway[s].items(): 
            if state not in explored: 
                explored.add(state) 
                path2 = path + [action, state] 
                if state == destination: 
                    return path2
                else: queue.append(path2)
    return []

                    
```

```python
global state 
# Set the initial state
state = INITIAL
global intent 

def respond(policy, message):
    global state 
    # Whether necessary parameters are provided
    if(state in [1, 2, 3, 4]):
        entities = pattern.findall(message)
        intent = "specify"
        
    # Plan a bus route   
    else: 
        if(state == 8): 
            bus_contained = pattern_bus.findall(message)
            if(len(bus_contained)>0):
                intent = "bus"
                
    # Acquire the intent
        else:
            intent = interpreter.parse(message)["intent"]["name"]
            
    # Change the state            
    (state, response_NO) = policy[(state, intent)]
    
    # If necessary parameters are provided, one-round query
    entities = pattern.findall(message)
    if(intent in ["time_table", "line_info", "station_info", "route_planning"] and len(entities) > 0):
        state = state + 4
        response_NO = response_NO + 4
        
    # Generate the response
    if response_NO in [10, 11, 12, 13, 14, 19, 20]:
        response = random.choice(responses[response_NO])
    else:
        if response_NO == 15:
            result = search_time_table(entities[0])
            results = list(result[0])
            response = 'The running time of Line {} is\n{}: {}\n{}: {} '.format(str(entities[0]), str(results[0]), str(results[2]), str(results[1]), str(results[3]))
       
        if response_NO == 16:
            result = search_line_info(entities[0])
            response = 'The route of Line {}:\n{}'.format(str(entities[0]), str(result[0][0]))
            
        if response_NO == 17:
            result = search_station_info(entities[0])
            response = 'You can take {} at {}'.format(str(result[0][0]),str(entities[0]))
            
        # Plan a subway route   
        if response_NO == 18:
            global starting_point 
            global destination 
            starting_point = entities[0]
            destination = entities[1]
            result = plan_path(starting_point, destination)
            path_output = organize_path(result)
            response = 'Here is the shortest path:\n'
            i = 1
            for line, stations_per_line in path_output.items():
                response = response + line + ': ' + '->'.join(stations_per_line) 
                if(i != len(path_output)):
                    response = response + '\n'
                i += 1
       
        if response_NO == 21:  
            response = plan_bus_path(starting_point, destination)
    
    return response
```

```python
# Iterate the subway route
def organize_path(result):
    elements_station = []
    elements_line = []
    i = 0
    while i < len(result):
        if i % 2 == 0:
            elements_station.append(result[i])
        else:
            elements_line.append(result[i])
        i += 1

    index_of_line = {}
    i = 0
    while i < len(elements_line):
        index_of_line[elements_line[i]] = i
        i += 1

    path_output = {}
    i = 0
    for line, index in index_of_line.items():
        stations_per_line = []
        while i <= index + 1:
            stations_per_line.append(elements_station[i])
            i += 1
        i -= 1
        path_output[line] = stations_per_line

    return path_output

```

```python
import requests
# plan bus route
def plan_bus_path(starting_point, destination):
    # Acquire the GPS coordinate
    url = 'http://api.map.baidu.com/geocoder/v2/'
    params_starting_point = {
                            'address': starting_point,
                            'city': '北京',
                            'output': 'json',
                            'ak': '2b4f75ed3baea23d71dc1828331f8dfc'
                            }
    params_destination = {
                         'address': destination,
                         'city': '北京',
                         'output': 'json',
                         'ak': '2b4f75ed3baea23d71dc1828331f8dfc'
                         }
    r = requests.get(url,params_starting_point) 
    r_js = r.json()
    longitude_starting_point = r_js['result']['location']['lng']
    latitude_starting_point = r_js['result']['location']['lat']
    r = requests.get(url,params_destination) 
    r_js = r.json()
    longitude_destination = r_js['result']['location']['lng']
    latitude_destination = r_js['result']['location']['lat']

    location_starting_point = str(latitude_starting_point)[0:9] + ',' + str(longitude_starting_point)[0:10]
    location_destination = str(latitude_destination)[0:9] + ',' + str(longitude_destination)[0:10]
    
    # Acquire the bus route
    url = 'http://api.map.baidu.com/direction/v2/transit'
    params = {
             'origin': location_starting_point,
             'destination': location_destination,
             'tactics_incity': '3',
             'output': 'json',
             'ak': '2b4f75ed3baea23d71dc1828331f8dfc'
             }
    r = requests.get(url,params)  
    r_js = r.json()
    steps = r_js['result']['routes'][0]['steps']
    i = 0
    
    # Generate the reply
    response = ''
    while i < len(steps):
        response = response + 'Step {}: '.format(i + 1) + str(steps[i][0]['instructions'])
        if(i != len(steps) - 1):
            response = response + '\n'
        i += 1
    return response
```
